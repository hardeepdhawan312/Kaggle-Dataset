```python
df=pd.read_csv('C:\\Users\\manish\\Documents\\Data_Science_Projects\\GeoMapsVisualisation\\2011_us_ag_exports.csv')
```

```python
df.head()
```

```python
data = dict(type='choropleth',
           colorscale='YIOrRd',
           locations=df['code'],
           locationmode='USA-states',
            text=df['state'],
           z= df['total exports'],
           colorbar={'title':'Millions USD'},
           marker=dict(line=dict(color = 'rgb(12,12,12)',width=1)))
```

```python
layout = dict(title='US Agriculture Reports 2011',
             geo = dict(scope='usa'))
```

```python
layout
```

```python
choromap= go.Figure(data=[data],layout=layout)
```

```python
iplot(choromap)
```

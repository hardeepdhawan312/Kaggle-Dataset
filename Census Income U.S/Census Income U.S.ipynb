```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('C:\\Users\\manish\\Documents\\Data_Science_Projects\\Census Income U.S\\adult.csv')
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
#changing categorical variables into numeric

dataset['income'] = dataset['income'].map({'<=50K': 0, '>50K': 1})
dataset['sex'] = dataset['sex'].map({'Male': 0, 'Female': 1})
dataset['race'] = dataset['race'].map({'Black': 0, 'Asian-Pac-Islander': 1, 'Other': 2, 'White': 3,'Amer-Indian-Eskimo': 4})
dataset['marital.status']=dataset['marital.status'].map({'Married-civ-spouse':2,'Divorced':4,'Never-married':5,' Separated':3,' Widowed':1,' Married-spouse-absent':0,'Married-AF-spouse':6})
dataset['occupation'] = dataset['occupation'].map({'?': 0, 'Farming-fishing': 1, 'Tech-support': 2,'Adm-clerical': 3, 'Handlers-cleaners': 4, 'Prof-specialty': 5,'Machine-op-inspct': 6, 'Exec-managerial': 7, 'Priv-house-serv': 8, 'Craft-repair': 9, 'Sales': 10,'Transport-moving': 11, 'Armed-Forces': 12, 'Other-service': 13,'Protective-serv': 14})
dataset['relationship'] = dataset['relationship'].map({'Not-in-family':0, 'Unmarried':0, 'Own-child':0, 'Other-relative':0, 'Husband':1, 'Wife':1})
dataset['education']= dataset['education'].map({'Bachelors':0,'Some-college':1,'11th':2,'HS-grad':3,' Prof-school':4,'Assoc-acdm':5,' Assoc-voc':6,'9th':7,' 7th-8th':8,'12th':9,'Masters':10,'1st-4th':11,'10th':12,'Doctorate':13,'5th-6th':14,'Preschool':15})
dataset['workclass']=dataset['workclass'].map({'Private':0, 'State-gov':1, 'Federal-gov':2, 'Self-emp-not-inc':3, 'Self-emp-inc':4, 'Local-gov':5, 'Without-pay':6,'?':7})
```

```python
# remove rows where occupation is unknown
dataset.occupation.replace(0, np.nan, inplace=True)
```

```python
dataset=dataset.dropna()
```

```python
dataset.head()
```

```python
#Labelling the dataset
X = dataset.loc[:,['age','fnlwgt','marital.status','education','education.num','occupation','relationship','race','sex','capital.gain','capital.loss','hours.per.week']]
y = dataset.loc[:,'income'] 
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Fitting Decision Tree Classification to the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
# Applying k-Fold Cross Validation on Decision Tree
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print(accuracies.mean())
print(accuracies.std())
```
